```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mousedf=pd.read_csv(mouse_drug_data_to_load)
clinicaldf=pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
df = pd.merge(clinicaldf,mousedf,how='left',on='Mouse ID')

# Display the data table for preview
df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
df1 = df.copy()
df1 = df1.groupby(['Drug','Timepoint'])
meantv=df1['Tumor Volume (mm3)'].mean()

# Convert to DataFrame
dfmean=pd.DataFrame(meantv)
dfmean.reset_index(level=1,inplace=True)
dfmean.reset_index(level=0,inplace=True)

# Preview DataFrame
```

```python
dfmean.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
stanerrtv=df1['Tumor Volume (mm3)'].sem()
# Convert to DataFrame
dfse=pd.DataFrame(stanerrtv)
dfse.reset_index(level=1,inplace=True)
dfse.reset_index(level=0,inplace=True)
# Preview DataFrame

```

```python
dfse.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
re_df=dfmean.pivot(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')
# Preview that Reformatting worked
```

```python
re_df.head()
```

```python
# Generate the Plot (with Error Bars)
#set variables from prev dataframe
y = re_df.index.tolist()
capomulin = re_df['Capomulin']
infubinol = re_df['Infubinol']
ketapril = re_df['Ketapril']
placebo = re_df['Placebo']
# plotting the lines
capomulin, = plt.plot(y,capomulin,color='red',alpha=0.7,label = 'Capomulin',marker='o')
infubinol, = plt.plot(y,infubinol,color='blue',alpha=0.7,label = 'infubinol',marker='^')
ketapril, = plt.plot(y,ketapril,color='green',alpha=0.7,label = 'ketapril',marker='s')
placebo, = plt.plot(y,placebo,color='black',alpha=0.7,label = 'placebo',marker='d')
#plot styling
plt.legend(handles=[capomulin,infubinol,ketapril,placebo], loc="best",fontsize='small')
plt.grid()
plt.ylabel('Tumor Volume(mm3)')
plt.xlabel('Time(Days)')
plt.title('Tumor Response to Treatment')

#show the plot
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
df2 = df.copy()
df2 = df2.groupby(['Drug','Timepoint'])
meanms=df1['Metastatic Sites'].mean()
# Convert to DataFrame
msmean=pd.DataFrame(meanms)

# Preview DataFrame
```

```python
msmean.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
stanerrms=df2['Metastatic Sites'].sem()
# Convert to DataFrame
msse=pd.DataFrame(stanerrms)
msse.reset_index(level=1,inplace=True)
msse.reset_index(level=0,inplace=True)


# Preview DataFrame
```

```python
msse.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
msmean.reset_index(level=1,inplace=True)
msmean.reset_index(level=0,inplace=True)
re_ms=msmean.pivot(index='Timepoint',columns='Drug',values='Metastatic Sites')
# Preview that Reformatting worked
re_ms.head()
```

```python
# Generate the Plot (with Error Bars)
ms_y = re_ms.index.tolist()
ms_capomulin = re_ms['Capomulin']
ms_infubinol = re_ms['Infubinol']
ms_ketapril = re_ms['Ketapril']
ms_placebo = re_ms['Placebo']
# plotting the lines
ms_capomulin, = plt.plot(ms_y,ms_capomulin,color='red',alpha=0.5,label = 'Capomulin',marker='o')
ms_infubinol, = plt.plot(ms_y,ms_infubinol,color='blue',alpha=0.5,label = 'infubinol',marker='^')
ms_ketapril, = plt.plot(ms_y,ms_ketapril,color='green',alpha=0.5,label = 'ketapril',marker='s')
ms_placebo, = plt.plot(ms_y,ms_placebo,color='black',alpha=0.5,label = 'placebo',marker='d')
#plot styling
plt.legend(handles=[ms_capomulin,ms_infubinol,ms_ketapril,ms_placebo], loc="best",fontsize='small')
plt.grid()
plt.ylabel('Met. Sites')
plt.xlabel('Treatment Duration(Days)')
plt.title('Metastatic Spread During Treatment')

#show the plot
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
df3 = df.copy()
df3 = df3.groupby(['Drug','Timepoint'])
mcount=df3['Mouse ID'].count()

# Convert to DataFrame
mcmean=pd.DataFrame(mcount)
mcmean.reset_index(level=1,inplace=True)
mcmean.reset_index(level=0,inplace=True)
mcmean=mcmean.rename(columns={'Mouse ID': 'Mouse Count'})


# Preview DataFrame
mcmean.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
re_mc=mcmean.pivot(index='Timepoint',columns='Drug',values='Mouse Count')
# Preview that Reformatting worked
re_mc.head()
```

```python
# Generate the Plot (Accounting for percentages)
mc_y = re_mc.index.tolist()
mc_capomulin = re_mc['Capomulin']
mc_infubinol = re_mc['Infubinol']
mc_ketapril = re_mc['Ketapril']
mc_placebo = re_mc['Placebo']
# plotting the lines
mc_capomulin, = plt.plot(mc_y,mc_capomulin*4,color='red',alpha=0.9,label = 'Capomulin',marker='o',linewidth='0.5')
mc_infubinol, = plt.plot(mc_y,mc_infubinol*4,color='blue',alpha=0.9,label = 'infubinol',marker='^',linewidth='0.5')
mc_ketapril, = plt.plot(mc_y,mc_ketapril*4,color='green',alpha=0.9,label = 'ketapril',marker='s',linewidth='0.5')
mc_placebo, = plt.plot(mc_y,mc_placebo*4,color='black',alpha=0.9,label = 'placebo',marker='d',linewidth='0.5')
#plot styling
plt.legend(handles=[mc_capomulin,mc_infubinol,mc_ketapril,mc_placebo], loc="best",fontsize='small')
plt.grid()
plt.ylabel('Survival Rate(%)')
plt.xlabel('Time(Days)')
plt.title('Survival During Treatment')

#show the plot
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
df4= re_df.copy()
summary_data = ((df4.iloc[-1]-df4.iloc[0])/df4.iloc[0])*100

# Display the data to confirm
summary_data
```

```python
# Store all Relevant Percent Changes into a Tuple
#create the dataframe for the summary bar chart.
drug_list=df4.columns.tolist()
summary_list = zip(drug_list,summary_data)
summarylist = []
for x in summary_list:
    summarylist.append(x)
summarylist
summary = pd.DataFrame(summarylist)
summary = summary=summary.rename(columns={0:'Drug',1:'Percentage Change'})
#create a column indicating whether percentage change is positive or negative.
summary['positive'] = summary['Percentage Change']>0
summary = summary.set_index('Drug')
#locate the four drug data we need for the summary chart
summary = summary.loc[['Capomulin','Infubinol','Ketapril','Placebo']]



#Plot the figure
plt.figure(figsize=(6,4))
fig = summary['Percentage Change'].plot(kind='bar',
                          color=summary.positive.map({True:'red',False:'green'}),width=-1,align='edge') 
fig.set_xlim(-1.2,len(summary.index.tolist())-0.8)
fig.set_ylim(-25,70)
fig.set_xticklabels(labels=summary.index.tolist(),rotation=0)
fig.set_ylabel('% Tumer Volumn Change')
fig.set_xlabel('')
fig.set_title('Tumor Change Over 45 Day Treatment')
fig.grid()

rects = fig.patches
summary['Percentage Change'] = summary['Percentage Change'].astype(float).map('{:,.0f}%'.format)
labels = summary['Percentage Change'].tolist()


#Use function to label the percentage change
for rect,label in zip(rects,labels):
    height = rect.get_height()
    if height > 0:
        label_position = 1
    else:
        label_position = -7
    fig.text(rect.get_x()+rect.get_width()/2,label_position,label,ha='center',va=
            'bottom',color = 'w')

# Show the Figure
```
